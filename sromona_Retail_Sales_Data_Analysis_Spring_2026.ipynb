### Question 4: Monthly Revenue Calculation (2 Marks)

Calculate the total revenue ('Total Amount') for each month. The result should be a new DataFrame or Series indexed by the month. Display the result.

**Expected Output:** A Series with month-year periods as the index and the corresponding total revenue as values.

```python
# Write your answer here
df_sales["Date"] = pd.to_datetime(df_sales["Date"])
df_sales["Month"] = df_sales["Date"].dt.month
df_sales["Year"] = df_sales["Date"].dt.year
df_sales["Month-Year"] = df_sales["Month"].astype(str) + "-" + df_sales["Year"].astype(str)
monthly_revenue = df_sales.groupby("Month-Year")["Total Amount"].sum()
print(monthly_revenue)
```

### Question 5: Weekly Revenue Calculation (1 Marks)

Similar to the last question, now calculate the total revenue ('Total Amount') for each week. Display the head of the resulting Series.

**Expected Output:** The first 5 rows of a Series with week periods as the index and total revenue as values.

```python
# Write your answer here
from google.colab import files
uploaded = files.upload()
import numpy as np
import pandas as pd
df_temp = pd.read_excel("retail_sales_dataset.xlsx")
df_temp.to_csv("retail_sales_dataset.csv", index=False)
df_sales = pd.read_csv("retail_sales_dataset.csv")
df_sales["Date"] = pd.to_datetime(df_sales["Date"])
df_sales["Week"] = df_sales["Date"].dt.isocalendar().week
df_sales["Year"] = df_sales["Date"].dt.year
df_sales["Week-Year"] = df_sales["Week"].astype(str) + "-" + df_sales["Year"].astype(str)
weekly_revenue = df_sales.groupby("Week-Year")["Total Amount"].sum()
print(weekly_revenue.head())
```

### Question 6: Best-Selling Product Category (2 Marks)

Determine which 'Product Category' has the highest total revenue. Group the data by 'Product Category', sum the 'Total Amount', and find the category with the maximum sum.

**Expected Output:** A printout of the name of the best-selling category and its total revenue.

### Question 7: Top 5 Customers by Spending (2 Marks)

Identify the top 5 customers who have spent the most in total. Group the data by 'Customer ID', sum their 'Total Amount', and display the top 5.

**Expected Output:** A Series showing the Customer IDs of the top 5 spenders and their total spending.

```python
# Write your answer here
top_5_customers = df_sales.groupby("Customer ID")["Total Amount"].sum().nlargest(5)
print(top_5_customers)
```

### Question 8: Average Order Value per Customer (2 Marks)

Calculate the average order value (AOV) for each customer. Group by 'Customer ID' and find the mean of the 'Total Amount'. Display the AOV for the top 5 customers by average spending.

**Expected Output:** A Series showing the Customer IDs of the top 5 customers by average order value and their AOV.

```python
# Write your answer here
top5_aov = df_sales.groupby('Customer ID')['Total Amount'].mean()
top5_aov = top5_aov.reset_index()
top5_aov.columns = ['Customer ID', 'AOV']

top5_aov['AOV'] = top5_aov['AOV'].round(2)
top5_aov = top5_aov.sort_values(by='AOV', ascending=False)
print("Top 5 Customers by Average Order Value (AOV):")
top5_aov.to_string(index=False)
print(top5_aov.head(5))
```

### Question 9: Cumulative Revenue Calculation (2 Marks)

Calculate and display the cumulative sum of revenue over time. First, group the data by 'Date' and sum the 'Total Amount'. Then, calculate the cumulative sum on this result. Display the last 5 days of the cumulative revenue series.

**Expected Output:** A Series showing the last 5 dates and the cumulative total revenue up to each of those dates.

```python
# Write your answer here
daily_revenue = df_sales.groupby('Date')['Total Amount'].sum()
cumulative_revenue = daily_revenue.cumsum()
cumulative_revenue.tail(5)
```

### Question 12: K-Means Clustering (5 Marks)

Import `KMeans` from `sklearn.cluster`. Initialize and fit the K-Means clustering model on the MNIST features (`X`).
Since we know there are 10 digits (0-9), set the number of clusters to 10. Assign the cluster labels to a variable `kmeans_labels`.

**Expected Output:** A successfully fitted K-Means model and the cluster labels array.

```python
from sklearn.cluster import KMeans
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# MNIST dataset loaded to define X and y
digits = load_digits()
X = digits.data
y = digits.target

# Initial K-Means Clustering
kmeans = KMeans(n_clusters=10, random_state=42)
kmeans.fit(X)
kmeans_labels = kmeans.labels_
kmeans_labels = kmeans.fit_predict(X)
print(kmeans)
print(kmeans_labels)
```

### Question 13: F1 Score Evaluation for Clustering (5 Marks)

Evaluate the clustering performance using the F1 score. Since K-Means assigns arbitrary cluster labels, you will first need to map each cluster label to the most frequent true label in that cluster.
After mapping the labels, calculate and print the macro-averaged F1 score using `sklearn.metrics.f1_score`.

**Expected Output:** The calculated F1 score of the clustering.

```python
# Write your answer here
import numpy as np
from scipy.stats import mode
from sklearn.metrics import f1_score, classification_report
y = y.astype(int)

mapped_labels = np.zeros_like(kmeans_labels)

for cluster in range(10):
    mask = (kmeans_labels == cluster)
    true_labels_in_cluster = y[mask]
    most_frequent = mode(true_labels_in_cluster)[0]
    mapped_labels[mask] = most_frequent
    print(f"Cluster {cluster} → Digit {most_frequent}")

f1 = f1_score(y, mapped_labels, average='macro')
print(f"\nMacro-Averaged F1 Score: {f1:.4f}")

print("\nDetailed Classification Report:")
print(classification_report(y, mapped_labels))
```
